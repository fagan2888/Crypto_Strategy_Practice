```python
def timestamp2date(timestamp):
    # function converts a Unix timestamp into Gregorian date
    return datetime.fromtimestamp(int(timestamp)).strftime('%Y-%m-%d')
```

```python
def fetchCryptoClose(fsym, tsym):
    # function fetches the close-price time-series from cryptocompare.com
    # it may ignore USDT coin (due to near-zero pricing)
    # daily sampled
    cols = ['date', 'timestamp', fsym]
    lst = ['time', 'open', 'high', 'low', 'close']
    timestamp_today = datetime.today().timestamp()
    curr_timestamp = timestamp_today
 
    for j in range(2):
        df = pd.DataFrame(columns=cols)
        url = "https://min-api.cryptocompare.com/data/histoday?fsym=" + fsym + \
              "&tsym=" + tsym + "&toTs=" + str(int(curr_timestamp)) + "&limit=2000"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        dic = json.loads(soup.prettify())
        for i in range(1, 2001):
            tmp = []
            for e in enumerate(lst):
                x = e[0]
                y = dic['Data'][i][e[1]]
                if(x == 0):
                    tmp.append(str(timestamp2date(y)))
                tmp.append(y)
            if(np.sum(tmp[-4::]) > 0):  # remove for USDT
                tmp = np.array(tmp)
                tmp = tmp[[0,1,4]]  # filter solely for close prices
                df.loc[len(df)] = np.array(tmp)
        # ensure a correct date format
        df.index = pd.to_datetime(df.date, format="%Y-%m-%d")
        df.drop('date', axis=1, inplace=True)
        curr_timestamp = int(df.ix[0][0])
        if(j == 0):
            df0 = df.copy()
        else:
            data = pd.concat([df, df0], axis=0)
    data.drop("timestamp", axis=1, inplace=True)
 
    return data  # DataFrame
```

```python
# N-Cryptocurrency Portfolio (tickers)
fsym = ['BTC', 'ETH', 'LTC', 'ARK']
        
#        'RDD', 'SC', 'DGB', 'ARK', 'STRAT', 'XRP', 'XVG', 'WINGS',
#        'DASH', 'WAVES', 'XMR', 'XRP', 'LTC', 'ETC', 'XEM', 'REP']
#         'MAID', 'ZEC', 'STEEM', 'GNT', 'FCT', 'ICN', 'DGD', 
#         'WAVES', 'DCR', 'LSK', 'DOGE', 'PIVX']
```

```python
tsym = 'USD'
```

```python
for e in enumerate(fsym):
    print('{}. {}'.format(e[0], e[1]))
    if(e[0] == 0):
        try:
            data = fetchCryptoClose(e[1], tsym)
        except:
            pass
    else:
        try:
            data = data.join(fetchCryptoClose(e[1], tsym))
        except:
            pass
```

### Get number of up and down moves

```python
df_eth['Regime'] = np.where(df_eth['Return'] < 0, -1, 0)
df_eth['Regime'] = np.where(df_eth['Return'] > 0, 1, df_eth['Regime'])

df_eth['Regime'].value_counts()
```

```python
today = datetime.today()
fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].scatter(df_eth.loc['2015-01-01':'2016-01-01'].index, df_eth.loc['2015-01-01':'2016-01-01']['Regime'])
axes[1].scatter(df_eth.loc['2016-01-01':'2017-01-01'].index, df_eth.loc['2016-01-01':'2017-01-01']['Regime'])
axes[2].scatter(df_eth.loc['2017-01-01':today].index, df_eth.loc['2017-01-01':today]['Regime'])

plt.tight_layout()
```

Number of down days high in first two years, but this year has very few.

```python
def get_value_counts_df(df, col):
    arr = []
    today = datetime.today()
    arr.append(df.loc['2015-01-01':'2016-01-01'][col].value_counts())
    arr.append(df.loc['2016-01-01':'2017-01-01'][col].value_counts())
    arr.append(df.loc['2017-01-01':today]       [col].value_counts())


    df_ret = pd.DataFrame(arr, 
                        index=['2015 - 2016', '2016 - 2017', '2017 - pres.'])
    df_ret['Total'] = df_ret.sum(axis=1)
    df_ret
    
    df_ret['-1 %'] = round(df_ret[-1] / df_ret['Total'] * 100, 2)
    df_ret['0  %'] = round(df_ret[0]  / df_ret['Total'] * 100, 2)
    df_ret['1  %'] = round(df_ret[1]  / df_ret['Total'] * 100, 2)
    
    return df_ret
```

```python
df_eth_regimes = get_value_counts_df(df_eth, 'Regime')
df_eth_regimes
```

More activity in 2016 and 2017, as expected. Majority up moves this year, 2017.

### Generate signal for ether off of t-1

```python
# shift(1) moves everything DOWN 1
# r_{t} - r_{t-1}
df_eth['Signal'] = np.sign(df_eth['Regime'] - df_eth['Regime'].shift(1))
df_eth['Signal'].plot(ylim=(-2,2))

df_eth['Signal'].value_counts()
```

```python
df_eth['Trade'] = np.where(df_eth['Signal'] == 1.0, 'Buy', 
                   np.where(df_eth['Signal'] == -1.0, 'Sell', 'None'))

df_eth['Trade'].value_counts()
```

```python
def Calc_Profit(date_start):
    df = df_eth.loc[date_start:]
    price_diff = df['Price'].values - df['Price'].shift(-1).values

    num_units = df.loc[df['Trade'] == 'Buy'].shape[0]

    df['Units'] = np.where(df['Trade'] == 'Buy', 1, 0).cumsum()

    df['Cost'] = np.where(df['Trade'] == 'Buy', df['Price'] * -1, 0)
    df['Cost_Cumm'] = df['Cost'].cumsum()

    df['Profit'] = (df['Units'] * df['Price']) - df['Cost']

    df[['Cost_Cumm', 'Profit']].plot(subplots=True, figsize=(6, 6), title=pd.to_datetime(date_start).year)
    
    return df
```

```python
df_eth_2016 = Calc_Profit('01-01-2016')
```

```python
df_eth_2016 = Calc_Profit('01-01-2017')
```

Total spent in 2016-Present is $1,000$ more ($3,500$ to $2,500$), but profit is $38,000$ more too ($50,000$ to $14,000$).
